```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
URL = "https://en.wikipedia.org/wiki/List_of_largest_cities"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")

results = soup.find(class_="plainrowheaders")

titles = soup.find_all('th', scope='col')
#print(titles)
#titles = soup.find_all('tr', class_='static-row-header')

titles_table = [title.text.strip() for title in titles]
columns = []
columns.append(titles_table[0])
columns.append(titles_table[1])
columns.append(titles_table[7])
columns.append(titles_table[8])
#for it in columns:
#   print(it)
df1 = pd.DataFrame(columns= columns)
```

```python
column_data = results.find_all('tr')
weather_city = []

for row in column_data[3:]:
    row_data = row.find_all(['th','td'])
    titles_table = [row_data[i].text.strip() for i in [0,1,4,5]]
    weather_city.append([row_data[i].text.strip() for i in [0]])
    #print(titles_table)
    df1.loc[len(df1)] = titles_table
df1
```
